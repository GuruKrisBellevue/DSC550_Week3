### 1.3. Use TextBlob to classify each movie review as positive or negative. Assume that a polarity score greater than or equal to zero is a positive sentiment and less than 0 is a negative sentiment.

```python
def calculatePolarity(input_blob):
    """
    This function is used to calculate the Polarity using TextBlob analyzer.
    It takes the input as the blob value and returns polarity of the sentiment.
    """
    polarity_1=input_blob.sentiment.polarity
    return polarity_1
```

```python
def calculateTextBlob_Sentiment(input_text):
    """
    This function is used to calculate the Sentiment and returns approriate binary values of sentiment
    based on the polarity. TextBlob analyzer is used in the function.
    """
    # Create a blob of the Input text
    blob_1=TextBlob(input_text)
    # Comptuting the polarity of the blob
    polarity_1=calculatePolarity(blob_1)
    # Assigning the default value of sentiment to be negative sentiment
    sentiment_1=0
    # If the polarity returned is greater than or equal to zero, it is taken as Positive sentiment
    if polarity_1 >= 0:
        sentiment_1=1
    # Return sentiment value
    return sentiment_1
```

```python
# Creating a variable for the review column
review_col=movie_review_df1["review"]
# Using apply to calculate the Sentiment values of all values of review column.
movie_review_df1["textBlob_sentiment"]=review_col.apply(calculateTextBlob_Sentiment)
```

```python
# Printing top 5 values of the movie review dataframe
movie_review_df1.head()
```

### 1.4 Check the accuracy of this model. Is this model better than random guessing?

```python
# Assigning the value of True Sentiment as the sentiment variable in the dataset
true_sent=movie_review_df1["sentiment"]
# The calculated sentiment using TextBlob analyzer is assigned to a variable 
textblob_sent=movie_review_df1["textBlob_sentiment"]
```

```python
def calculate_ModelAccuracy(true_sent,pred_sent,analyzer_name):
    """
    This function takes the input of True sentiment values and the calculated sentiment values and
    computes the accuracy of the model. The formula used in accuracy calulation is:
    Accuracy=((TP+TN)/(TP+TN+FP+FN))*100
    """
    truePos = np.sum(np.logical_and(pred_sent == 1, true_sent == 1))
    trueNeg = np.sum(np.logical_and(pred_sent == 0, true_sent == 0))
    falsePos = np.sum(np.logical_and(pred_sent == 1, true_sent == 0))
    falseNeg = np.sum(np.logical_and(pred_sent == 0, true_sent == 1))
    print(f"""
    Results of {analyzer_name} Analyzer:
        True positive is {truePos}
        True Negative is {trueNeg}
        False Positive is {falsePos}
        False Negative is {falseNeg}
    """)
    accuracy= ((truePos+trueNeg)/(truePos+trueNeg+falsePos+falseNeg))*100
    print(f"Accuracy of the model using  {analyzer_name} analyzer is {accuracy} %")
```

```python
calculate_ModelAccuracy(true_sent,textblob_sent,'TextBlob')
```

```python
# Using Random Guess Model to calculate the sentiment
# Setting seed to get same random results during each run
np.random.seed(100)
# Randomly guessing the sentiment either as 0 or 1 from 25000 samples and assigneing results to variable
random_guess=np.random.choice(2,len(true_sent))
# Printing sample results
random_guess[:10]
```

```python
# Calculate the accuracy of the Random guess model
calculate_ModelAccuracy(true_sent,random_guess,'Random Choice')
```

```python
def calculate_VADER_sentiment(input_text):
    """
    This function calculates the sentiment using VADER analyzer. It takes the input text and 
    computes the sentiment based on the compound value returned by the Vader analyzer
    """
    # Generating the sentiment dictionary of the input text using Vader object
    sentiment_dict=vader_obj.polarity_scores(input_text)
    # Assigning default value of the sentiment
    sentiment_2=0
    # Is the compound value is greater than or equal to zero, it is positive sentiment.
    if sentiment_dict['compound']>=0:
        sentiment_2=1
    # Return the calculated sentiment values
    return sentiment_2
```

### 2.2.Remove punctuation and special characters from the text.

```python
"""
Creating a dictionary of the Punctuations. This is done by checking the Category of all the 
Unicode characters and filter the ones that starts with 'P'. Then Dictionary of the key:values is
created by assigning 'None' as the value.
"""
punctuation_dict = dict.fromkeys(item for item in range(sys.maxunicode)
                    if unicodedata.category(chr(item)).startswith('P'))
```

```python
def remove_punctuations(input_text):
    """
    This function takes the input text and removes the punctuations and returns the cleansed value
    """
    # Translate method is used to replace the actial value with None when a punction is encountered
    cleansed_list1 = [string.translate(punctuation_dict) for string in input_text]
    return cleansed_list1
```

```python
# Creating a list of all possible special characters that should be cleansed from the input
spl_char_list = [
    ',', '.', '"', ':', ')', '(', '!', '?', '|', ';', "'", '$', '&',
    '/', '[', ']', '>', '%', '=', '#', '*', '+', '\\', '•',  '~', '@', '£',
    '·', '_', '{', '}', '©', '^', '®', '`',  '<', '→', '°', '€', '™', '›',
    '♥', '←', '×', '§', '″', '′', 'Â', '█', '½', 'à', '…', '“', '★', '”',
    '–', '●', 'â', '►', '−', '¢', '²', '¬', '░', '¶', '↑', '±', '¿', '▾',
    '═', '¦', '║', '―', '¥', '▓', '—', '‹', '─', '▒', '：', '¼', '⊕', '▼',
    '▪', '†', '■', '’', '▀', '¨', '▄', '♫', '☆', 'é', '¯', '♦', '¤', '▲',
    'è', '¸', '¾', 'Ã', '⋅', '‘', '∞', '∙', '）', '↓', '、', '│', '（', '»',
    '，', '♪', '╩', '╚', '³', '・', '╦', '╣', '╔', '╗', '▬', '❤', 'ï', 'Ø',
    '¹', '≤', '‡', '√', '«', '»', '´', 'º', '¾', '¡', '§', '£', '₤']
# Creating a dictionary of special characters from the list created above. By default None will be
# assigned as the value
spl_char_dict=dict.fromkeys(spl_char_list)
spl_char_dict
```

```python
def remove_spl_chars(input_text):
    """
    This function takes the input text and removes the special characters and returns the cleansed value
    """
    # Translate method is used to replace the actial value with None when a special char is encountered
    cleansed_list2 = [string.translate(spl_char_dict) for string in input_text]
    return cleansed_list2
```

```python
# creating a list of all reviews 
txt1=movie_review_df2["review"].to_list()
# apply the function to remove punctuations to all the values in reviews column and create a new column
movie_review_df2["cleansed_review"]=remove_punctuations(txt1)
```

```python
def remove_html_tags(input_text):
    """Function to remove html tags from a string"""
    # Creating a pattern to remove all html tags and values within the tag
    clean_pattern = re.compile('<.*?>')
    # Cleansing the input string by replacing the HTML tag and its contents with ''
    cleansed_val= re.sub(clean_pattern, '', input_text)
    return cleansed_val
```

```python
# Applying the Remove HTML tag function for all values in the cleansed review column
movie_review_df2["cleansed_review"]=movie_review_df2["cleansed_review"].apply(remove_html_tags)
```

### 2.3. Remove stop words.

```python
# Importing the libraries required 
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
# Downloading the stopwords from nltk
nltk.download('stopwords')
# creating list of stop words in English language
stopwords_list=stopwords.words("english")
```

```python
# Printing top 10 values in the stopwords list
stopwords_list[:10]
```

```python
def remove_stop_words(input_text):
    """
    Function to remove stop words from the input text.
    Returns cleaner version after removing the stop words.
    """
    # Tokenizing the input text by creating token of each word
    tokenized_string=word_tokenize(input_text)
    # Using Filter and Lambda to create a filter object that contains all words 
    # except the ones in stopwords list
    filtered_words = filter(lambda token: token not in stopwords_list, tokenized_string)
    # Joining the words with spaces from the filtered list and returning the value
    joined_words = " ".join(filtered_words)
    return joined_words
```

```python
# Creating a new column noStopWords_review by applying the function all values of cleansed review column
movie_review_df2["noStopWords_review"]=movie_review_df2["cleansed_review"].apply(remove_stop_words)
```

```python
# Printing top 2 rows from dataframe including the noStopWords review column
movie_review_df2.head(2)
```

### 2.4. Apply NLTK’s PorterStemmer.

```python
# Downloading the libraries required for stemmming
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from functools import reduce
# Creating a stem object
port_stem = PorterStemmer()
```

```python
def porter_stemmer(input_text):
    """
    This function takes  the input text and returns a list of stemmed words of each word in the text.
    """
    # Tokenizing the Input text by using the word tokenize from nltk
    tokenized_string=word_tokenize(input_text)
    # Using reduce and lambda to apply the stem object on each Tokenized word and combine them. 
    stemmed_words = reduce(lambda word1, word2: word1 + " " + port_stem.stem(word2), tokenized_string, "")
    return stemmed_words
```

```python
# Creating a new column stemmer_review by applying the function all values of noStopWords review column
movie_review_df2["stemmer_review"]=movie_review_df2["noStopWords_review"].apply(porter_stemmer)
```

```python
# Printing top 2 rows from dataframe including the stemmer review column
movie_review_df2.head(2)
```
